```python
import numpy as np

# Define points
points = {
    "A": np.array([2, 3]),
    "B": np.array([3, 4]),
    "C": np.array([8, 7]),
    "D": np.array([9, 8]),
    "E": np.array([10, 9])
}

# Initial centroids
C1 = np.array([2, 3])
C2 = np.array([10, 9])

# Euclidean distance function
def euclidean(p, c):
    return np.sqrt(np.sum((p - c) ** 2))

# Assign points to nearest centroid
clusters = {1: [], 2: []}

for name, point in points.items():
    d1 = euclidean(point, C1)
    d2 = euclidean(point, C2)

    # Assign to nearest centroid
    if d1 < d2:
        clusters[1].append(point)
    else:
        clusters[2].append(point)

# Compute new centroids
new_C1 = np.mean(clusters[1], axis=0)
new_C2 = np.mean(clusters[2], axis=0)

print("New Centroid 1:", new_C1)
print("New Centroid 2:", new_C2)
```

```python
# Manhattan distance function
def manhattan(p, c):
    return np.sum(np.abs(p - c))

for name, point in points.items():
    print(
        name,
        manhattan(point, C1),
        manhattan(point, C2)
    )
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Convert points to array
X = np.array(list(points.values()))

# Single linkage
single = linkage(X, method='single')

# Complete linkage
complete = linkage(X, method='complete')

# Plot dendrograms
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
dendrogram(single)
plt.title("Single Linkage")

plt.subplot(1, 2, 2)
dendrogram(complete)
plt.title("Complete Linkage")
plt.show()
```
