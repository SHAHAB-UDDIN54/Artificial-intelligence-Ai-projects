```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Step 1: Load and Clean the Data**

```python
dataset = pd.read_csv("/content/house price.csv")
dataset.head(5)
```

**Step 2: Train the Models**
We train all three models to see who is the "winner."

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_absolute_error
```

**1. Data Loading aur Selection**Aapne computer ko bataya ke "Bedrooms" hamara sawal (input) hai aur "Price" hamara jawab (target) hai jo humne seekhna hai.

```python
x = dataset[["bedrooms"]]  # features
y = dataset[["price"]]      # Target
```

**2. Train-Test Split (Data ki Taqseem)**
Ye step boht zaroori hai. Humne 80% data model ko parhne ke liye diya, aur 20% data side par rakh diya taake baad mein check kar sakein ke model ne sahi seekha hai ya sirf ratta (memorize) lagaya hai.

```python
## Split data: 80% for training, 20% for testing
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

# Initialize Models
lr = LinearRegression()
ridge = Ridge(alpha = 0.1)
lasso = Lasso(alpha = 0.1)

# Fit Models (Training)
lr.fit(x_train, y_train)
ridge.fit(x_train,y_train)
lasso.fit(x_train,y_train)
```

**Step 3: Compare MAE (Mean Absolute Error)**
MAE tells us: "On average, how many dollars ($) is our prediction off?" Lower MAE is bette

```python
# Make Predictions
pred_lr = lr.predict(x_test)
pred_ridge = ridge.predict(x_test)
pred_lasso = lasso.predict(x_test)

# Calculate MAE
mae_lr = mean_absolute_error(y_test, pred_lr)
mae_ridge = mean_absolute_error(y_test, pred_ridge)
mae_lasso = mean_absolute_error(y_test, pred_lasso)

print(f"Linear MAE: {mae_lr}")
print(f"Ridge MAE: {mae_ridge}")
print(f"Lasso MAE: {mae_lasso}")
```
