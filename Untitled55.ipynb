```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_excel("/content/hospital_patient_dataset_50.xlsx")
print(dataset.head(5))
```

checking the missing value


```python
print('Missing value in each column')
print(dataset.isnull().sum())
```

```python
dataset['Age'] = dataset['Age'].fillna(dataset['Age'].mean())
dataset['Gender'] = dataset['Gender'].fillna(dataset['Gender'].mode()[0])
dataset['Department'] = dataset['Department'].fillna(dataset['Department'].mode()[0])
dataset['Doctor'] = dataset['Doctor'].fillna(dataset['Doctor'].mode()[0])
dataset['Discharge_Date'] = dataset['Discharge_Date'].fillna(dataset['Discharge_Date'].mode()[0])
dataset['Diagnosis'] = dataset['Diagnosis'].fillna(dataset['Diagnosis'].mode()[0])
```

```python
dataset.isnull().sum()
```

```python
dataset.replace([np.inf,-np.nan],np.nan,inplace=True)
```

remove the duplicates value


```python
dataset.drop_duplicates(inplace=True)
```

replace negative salaries

```python
dataset['Room_No'] = np.where(dataset['Room_No'] < 0, dataset['Room_No'].mean(), dataset['Room_No'])
```

```python
dataset.to_csv('cleaned_hosptil_patient_data.csv',index=False)
```
