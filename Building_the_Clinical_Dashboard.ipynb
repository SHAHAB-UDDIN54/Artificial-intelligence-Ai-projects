```python
import numpy as np
```

Step 1 — Import numpy

Numpy helps us do math easily (like average and standard deviation)

patient data

```python
data = [4.2, 5.1, 3.8, 7.2, 5.5, 6.0]
```

Step 2 — Create the function

This function:
• Takes in a list of numbers
• Finds the mean (average)
• Finds the standard deviation (how spread out the numbers are)
• Returns both

```python
def summarize_clinic(data):
    mean = np.mean(data)       # Finds the average
    std_dev = np.std(data)    # Finds the standard deviation
    return mean, std_dev
```

Step 3 — Run the function

This prints the results

```python
mean, std = summarize_clinic(data)

print("Mean:", mean)
print("Standard Deviation:", std)
```

Step 4 — Add the outlier (15.0)

This line adds a new patient score to the lis

```python
data.append(15.0)
```

Now run the function again

```python
mean, std = summarize_clinic(data)

print("New Mean:", mean)
print("New Standard Deviation:", std)
```

Example Dosage vs Recovery Da

```python
dosage = [10, 20, 30, 40, 50]
recovery = [1, 2, 3, 4, 5]
```

Step 1 — Import matplotlib

This lets us draw graphs

```python
import matplotlib.pyplot as plt
```

Step 2 — Scatter Plot (Dots)

This shows the real data points

```python
plt.scatter(dosage, recovery)
```

Step 3 — Prediction Line (w = 0.1)

We use this formula:

y = w × x

```python
w = 0.1
predicted = [w * x for x in dosage]

plt.plot(dosage, predicted, label="w = 0.1")
```

Step 4 — Plot Multiple Lines (FOR LOOP)

This draws 3 different prediction lines

```python
weights = [0.2, 0.5, 0.8]

for w in weights:
    predicted = [w * x for x in dosage]
    plt.plot(dosage, predicted, label=f"w = {w}")
```

Step 5 — Make It Look Nice

```python
plt.xlabel("Dosage")
plt.ylabel("Recovery")
plt.title("Dosage vs Recovery Prediction Lines")
plt.legend()
plt.show()
```
